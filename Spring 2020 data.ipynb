```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
#trying something out with pd
csv_initial = pd.read_csv("2020_spring.csv")
print(csv_initial.columns)
print(csv_initial.head())
csv_initial['league']
```

```python
data = []
with open('2020_spring.csv') as file:
    reader = csv.reader(file)
    data = list(reader)
header = data[0]
header
```

```python
data.pop(0)
# print(data[0])
lpl_101 = []
for i in data:
    if i[header.index('league')] == 'LPL':
        if i[header.index('patch')] == '10.01':
            if i[header.index('playerid')] == '100' or i[header.index('playerid')] == '200':
                lpl_101.append(i)
lpl_101   
```

```python
ban_count = {}
for i in lpl_101:
    if i[header.index('ban1')] in ban_count.keys():
        ban_count[i[header.index('ban1')]] += 1
    else:
        ban_count[i[header.index('ban1')]] = 1
    if i[header.index('ban2')] in ban_count.keys():
        ban_count[i[header.index('ban2')]] += 1
    else:
        ban_count[i[header.index('ban2')]] = 1
    if i[header.index('ban3')] in ban_count.keys():
        ban_count[i[header.index('ban3')]] += 1
    else:
        ban_count[i[header.index('ban3')]] = 1
    if i[header.index('ban4')] in ban_count.keys():
        ban_count[i[header.index('ban4')]] += 1
    else:
        ban_count[i[header.index('ban4')]] = 1
    if i[header.index('ban5')] in ban_count.keys():
        ban_count[i[header.index('ban5')]] += 1
    else:
        ban_count[i[header.index('ban5')]] = 1
lpl_101_matches = len(lpl_101)/2     
ban_count
```

```python
ban_percent = {}
for i in ban_count:
    if (ban_count[i] / lpl_101_matches)*100 > 20:
        ban_percent[i] = (ban_count[i] / lpl_101_matches)*100
ban_percent
```

```python
def ban_percent(region, patch, percent):
    region_patch = []
    for i in data:
        if i[header.index('league')] == region:
            if i[header.index('patch')] == patch:
                if i[header.index('playerid')] == '100' or i[header.index('playerid')] == '200':
                    region_patch.append(i)
    ban_count = {}
    for i in region_patch:
        if i[header.index('ban1')] in ban_count.keys():
            ban_count[i[header.index('ban1')]] += 1
        else:
            ban_count[i[header.index('ban1')]] = 1
        if i[header.index('ban2')] in ban_count.keys():
            ban_count[i[header.index('ban2')]] += 1
        else:
            ban_count[i[header.index('ban2')]] = 1
        if i[header.index('ban3')] in ban_count.keys():
            ban_count[i[header.index('ban3')]] += 1
        else:
            ban_count[i[header.index('ban3')]] = 1
        if i[header.index('ban4')] in ban_count.keys():
            ban_count[i[header.index('ban4')]] += 1
        else:
            ban_count[i[header.index('ban4')]] = 1
        if i[header.index('ban5')] in ban_count.keys():
            ban_count[i[header.index('ban5')]] += 1
        else:
            ban_count[i[header.index('ban5')]] = 1
    region_patch_matches = len(region_patch)/2   
    ban_percent = {}
    for i in ban_count:
        if (ban_count[i] / region_patch_matches)*100 > percent:
            ban_percent[i] = (ban_count[i] / region_patch_matches)*100
    keys = tuple(ban_percent.keys())
    y_pos = np.arange(len(keys))
    values = list(ban_percent.values())
    plt.figure(figsize=(15,10)) 
    plt.barh(y_pos, values)
    plt.yticks(y_pos, keys)
    plt.title(region + " " + patch + ' Ban Presence > ' + str(percent))
    plt.xlabel("Percent Presence")
    plt.ylabel("Champions")

    plt.show()
#     return ban_percent
ban_percent('LPL','10.06', 30)
ban_percent('LEC','10.06', 30)
```
